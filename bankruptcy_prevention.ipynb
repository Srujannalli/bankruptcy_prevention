```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.20,random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()
```

## LOGISTIC REGRSSION MODEL

```python
logistic_model.fit(X_train,Y_train)
```

```python
logistic_model.coef_ 
```

```python
logistic_model.score(X_test,Y_test)
```

```python
logistic_model.predict_proba(X)
```

```python
logistic_pred = logistic_model.predict(X)
logistic_pred
```

## DECISION TREE MODEL

```python
from sklearn import tree
```

```python
decision_tree_model = tree.DecisionTreeClassifier(criterion="entropy", max_depth=3)
```

```python
decision_tree_model.fit(X_train, Y_train)
```

```python
tree.plot_tree(decision_tree_model);
```

```python
fn=['industrial_risk','management_risk','financial_flexibility','credibility','competitiveness','operating_risk']
cn=['bankruptcy', 'non-bankruptcy']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(decision_tree_model,
               feature_names = fn, 
               class_names=cn,
               filled = True);
```

```python
decision_pred = decision_tree_model.predict(X)
decision_pred
```

```python
pd.Series(decision_pred).value_counts()
```

```python
decision_tree_model.score(X_test, Y_test)
```

## XGBOOST CLASSIFIER MODEL

```python
!pip install xgboost
from xgboost import XGBClassifier
```

```python
xg_model = XGBClassifier(max_depth =6, n_estimators=500, learning_rate=0.50,gamma=0.5, objective='binary:logistic')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y_xg_train = le.fit_transform(Y_train)
Y_xg_test = le.fit_transform(Y_test)
xg_model.fit(X_train,Y_xg_train)
```

```python
xg_pred = xg_model.predict(X_test)
xg_pred
```

```python
predictions = [round(value) for value in xg_pred] 
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(predictions,Y_xg_test)
print(f"Accuracy : {accuracy*100}%")
```

```python
xg_model.score(X_test,Y_xg_test)
```

# MODEL VALIDATION 

```python
from sklearn.model_selection import cross_val_score
```

```python
k_scores = []
models = [logistic_model,decision_tree_model,rf_model,svm_model,nb_model,knn_model]
for x in range(0,len(models)):
    scores = cross_val_score(models[x],X,Y)
    k_scores.append(scores.mean())
```

```python
k_scores
```

```python
xg_score = cross_val_score(xg_model,X_train,Y_xg_train)
k_scores.append(xg_score.mean())
```

```python
k_scores
```

```python
for x in range(1,len(k_scores)):
    print(f"{models[x-1]} accuracy : {k_scores[x-1]}")
```
